```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.svm import LinearSVC
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

c = [0.001, 0.01, 0.05, 0.1, 0.5, 0.75, 1]
maxft=[1000, 5000, 10000, 50000, 100000, 120000, 140000, 160000, 180000, 200000, 220000, 250000]
for j in maxft:
  x1=x1new
  
  vectorizer = TfidfVectorizer(stop_words = 'english', max_features=j) #, ngram_range=(1, 2))
  x1 = vectorizer.fit_transform(x1)
  X1_train, X1_test, y1_train, y1_test = train_test_split(x1, y, test_size=0.20, random_state=42)
  print(j)
  for i in c:
      model_lr = LogisticRegression(C = i, max_iter = 1000)
      model_lr.fit(X1_train, y1_train)
      y_pred = model_lr.predict(X1_test)
      accuracy_lr = accuracy_score(y1_test, y_pred)
      # f1_lr = f1_score(y_test, y_pred, average='micro')
      print ("Logistic C = ", i, " Accuracy: ", accuracy_lr)
      # print ("C = ", i, " F1 score: ", f1_lr)
      model_svm = LinearSVC(C=i)
      model_svm.fit(X1_train, y1_train)
      y_pred = model_svm.predict(X1_test)
      accuracy_svm = accuracy_score(y1_test, y_pred)
      # f1_svm = f1_score(y_test, y_pred)
      print ("SVM C = ", i, " Accuracy: ", accuracy_svm)
      # print ("C = ", i, " F1 score: ", f1_svm)
```
